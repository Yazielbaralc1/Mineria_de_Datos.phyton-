
*   Descrpcion básica de columnas (tipo, elementos, celdas vacias, etc.)


```python
print("El número de columnas en la base datos Data.csv es:\n",len(df.columns.values))
print("El número en la base datos Data.csv es:\n",len(df.index))
print("\nNombre de columna / Tipo de dato\n")
print(df.dtypes,"\n")  #dtypes nos muestra por columna que tipo de dato tiene cada uno
print("Nombre de columna |  Datos nulos\n")
df.isnull().sum() #isnull nos ubica que datos son nulos junto con .sum que esto nos suma los datos nulos indicados
```



*   El objetivo es una clasificación con los datos, asi que hay que acomodar los datos de manera que se nos facilite esta tarea.




```python
print(df.dtypes.unique())
#Acomodamos por columnas el tipo de dato que representara
df=df.reindex(columns=['points','price','country','province'])
df #Imprimos los datos reacomodados para una mejor lectura, para tener el objetivo y manipular los datos de manera que podemos analizar un costo beneficio dependiendo  de cada pais con su provincia.
```




*   Cambiar nombres de columnas.
