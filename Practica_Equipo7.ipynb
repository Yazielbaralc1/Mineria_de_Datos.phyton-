* Objetivo: Ayudar a terceros a mejorar sus ventas, buscando la mejor zona para empresas de diferentes giros que podrían sacar algún beneficio del flujo y tipo de empleados de acuerdo con las necesidades de personal de AMAZON.

```python
print(db1.dtypes.unique())
#Reordenamos las columnas
db1=db1.reindex(columns=['Title','location','Posting_date'])
db1 #Se muestran los datos ordenados, de tal manera que es más fácil ver la relación vacante-lugar
```

* Renombrar columnas

```python
db1=db1.rename(columns={'Title':'Vacante','location':'Ciudad','Posting_date':'Publicación'})
db1
```

```python
db1['Ciudad'].value_counts().plot.bar()
```

En la gráfica anterior podemaos ver el número de vacantes para cada ciudad del mundo, sin embargo los datos no se logran a preciar correctamente. 

```python
db1['Ciudad'].value_counts().head(10).plot.bar()
```

En esta gráfica al disminuir el numero de cuidades mostradas se logra apreciar mejor el número de vacantes para cada ciudad, notando que la ciudad con mayor numero de vacantes es US, WA, Seattle con poco más de 1750 vacantes.

```python
(db1['Ciudad'].value_counts().head(10) / len(db1)).plot.bar()
```

En esta gráfica se muestra en porcentanje de vacantes para cada ciudad, pese a que US, WA, Seattle es la cuidad con el mayor número de vacntes representa poco más del 0.5%.

```python
db1['Publicación'].value_counts().sort_index().plot.bar()
```

Esta gráfica muestra la distribución de las vacantes de acuerdo a su publicación, sin embargo al ser muchas las fechas de publicación, estás no se identifican correctamente, aun así se logra apreciar que el mayor número de vacantes que se publicaron en un mismo día es poco mas de 50.
